```python
import numpy as np
import matplotlib.pyplot as plt

# Sample Data
initial_scores = np.array([84, 78, 83, 81, 81, 80, 79, 85, 76, 83])
final_scores = np.array([86, 86, 80, 86, 84, 86, 86, 82, 83, 84])

# Bootstrapping
np.random.seed(0)  # for reproducibility
differences = final_scores - initial_scores
boot_means = []
for _ in range(10000):
    sample = np.random.choice(differences, size=len(differences), replace=True)
    boot_means.append(np.mean(sample))

# Confidence Interval
CI = np.percentile(boot_means, [2.5, 97.5])
print(f"95% Confidence Interval: {CI}")
```

```python
plt.hist(boot_means, bins=30, edgecolor='black')
plt.axvline(CI[0], color='red', linestyle='dashed', linewidth=1)
plt.axvline(CI[1], color='red', linestyle='dashed', linewidth=1)
plt.title('Bootstrap Distribution of Mean Differences')
plt.xlabel('Mean Difference')
plt.ylabel('Frequency')
plt.show()
```
